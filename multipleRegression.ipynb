```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("Book2.csv")
import math
med=math.floor(df.bedroom.median())
print(med)
df=df.fillna(value={"bedroom":med})
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
```

```python
# reg.predict([[2700,3.0,14]])
ar=int(input("Enter area"))
bd=int(input("Enter number of bedroom"))
ag=int(input("Enter age"))
pr=reg.predict([[ar,bd,ag]])
```

```python
d=pd.read_csv("Book2.csv")
d['area']=pd.Series(ar)
d['bedroom']=pd.Series(bd)
d['age']=pd.Series(ag)
d['price']=pd.Series(pr)
d=d.dropna()
d.to_csv("Book2.csv",mode='a',index=False,header=False)
d
```
